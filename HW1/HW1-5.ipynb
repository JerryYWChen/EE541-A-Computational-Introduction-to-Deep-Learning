HW1-5(a)

```python
import matplotlib.pyplot as plt
import random
```

```python
trials = 50 # repeate 50 trials.
prob_head = 0.7 #set probablility of head.
longest_run = 0
current_run = 0
Tossing = random.choices(['H','T'], weights = [prob_head, 1 - prob_head], k=trials) #return a list with weights parameter  
num_heads = 0
for toss in Tossing:
    if toss == 'H':
        num_heads += 1
        current_run += 1
        if current_run > longest_run:
            longest_run = current_run
    else:
        current_run = 0
print('total number of heads:', num_heads)
print('the longest run:', longest_run)
```

HW1-5(b)

```python
import matplotlib .pyplot as plt
import random
```

```python
trials = 50 # repeate 50 trials.
prob_head = 0.70 #set probablility of head.
rep_experiments = [20, 100, 200, 1000] #repeat experiment list.

num_heads = 0
for r in rep_experiments:
    result = []
    for toss in range(r):
        Tossing = random.choices(['H','T'], weights = [prob_head, 1 - prob_head], k=trials) #return a list with weights parameter.
        num_heads = Tossing.count('H')
        result.append(num_heads) #add to result.
    #plot histogram
    plt.hist(result, bins = range(trials+2), edgecolor='black', label=f'{r} experiments')
    plt.title('Histogram of number of heads in 50 trials')
    plt.xlabel('Number of heads in 50 trials')
    plt.ylabel('Occurrence times')
    plt.legend()
    plt.show()
  
```
