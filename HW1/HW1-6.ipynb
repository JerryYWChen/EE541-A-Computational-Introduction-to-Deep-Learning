```python
import random
import matplotlib.pyplot as plt
```

```python
n_realizations= [100, 1000, 10000] #list of different realization

for n in n_realizations:
    result = []
    for i in range(n):
        total=0
        count=0
        while total <= 4:
            total += random.uniform(0, 1) #get the random number between 0 and 1
            count += 1
        result.append(count)
    #plot the histogram
    plt.hist(result, bins = range(min(result), max(result)+2), alpha=0.7, edgecolor='black')
    plt.title(f'Histogram of N for {n} realizations')
    plt.xlabel('N')
    plt.ylabel('Occurrence times')
    plt.show()
```
